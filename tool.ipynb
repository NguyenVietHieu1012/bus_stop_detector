```python
import cv2

from ultralytics import YOLO

# Load the YOLO model
model = YOLO("./models/trained_model_tuned.pt")

# Open the video file
video_path = "reality_test/test2.mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLO inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLO Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

```python
import os
import shutil

# Đường dẫn tới thư mục chứa file gốc và thư mục đích
source_folder = '/path/to/source'
destination_folder = '/path/to/destination'

# Danh sách các file cần chuyển
video_files = [
    'FILE250221-161530-027100-M.MP4',
    'FILE250221-161630-027101-M.MP4',
    'FILE250221-161730-027102-M.MP4',
    'FILE250221-162131-027106-M.MP4',
    'FILE250221-162332-027108-M.MP4',
    'FILE250221-162833-027113-M.MP4',
    'FILE250221-163735-027122-M.MP4',
    'FILE250221-163835-027123-M.MP4',
    'FILE250221-163935-027124-M.MP4'
]

# Tạo thư mục đích nếu chưa tồn tại
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Di chuyển từng file từ thư mục nguồn sang thư mục đích
for video in video_files:
    source_path = os.path.join(source_folder, video)
    destination_path = os.path.join(destination_folder, video)
    if os.path.exists(source_path):
        shutil.move(source_path, destination_path)
        print(f"Đã chuyển file {video} sang {destination_folder}")
    else:
        print(f"File {video} không tồn tại trong thư mục nguồn")
```
