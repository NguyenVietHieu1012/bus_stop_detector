```python
import os
import shutil

# Đường dẫn tới thư mục chứa hình ảnh và nhãn
dataset_path = 'C:/Users/Admin/PycharmProjects/GamePro/bus_stop_detector/dataset'
images_path = os.path.join(dataset_path, 'images')
labels_path = os.path.join(dataset_path, 'labels')

# Đường dẫn tới thư mục đầu ra
output_path = 'C:/Users/Admin/PycharmProjects/GamePro/bus_stop_detector/split_dataset'

# Tạo thư mục đầu ra nếu chưa tồn tại
for split in ['train', 'test', 'val']:
    os.makedirs(os.path.join(output_path, split, 'images'), exist_ok=True)
    os.makedirs(os.path.join(output_path, split, 'labels'), exist_ok=True)

# Lấy danh sách tệp ảnh và sắp xếp chúng
image_files = sorted(f for f in os.listdir(images_path) if f.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')))
label_files = sorted(f for f in os.listdir(labels_path) if f.endswith('.txt'))

# Hàm để di chuyển tệp
def move_files(files, src_folder, dest_folder, split_type):
    for file in files:
        src = os.path.join(src_folder, file)
        dest = os.path.join(dest_folder, split_type, os.path.basename(src_folder), file)
        shutil.move(src, dest)

# Chia dữ liệu
move_files(image_files[:70], images_path, output_path, 'test')
move_files(label_files[:70], labels_path, output_path, 'test')

move_files(image_files[70:140], images_path, output_path, 'val')
move_files(label_files[70:140], labels_path, output_path, 'val')

move_files(image_files[140:], images_path, output_path, 'train')
move_files(label_files[140:], labels_path, output_path, 'train')

print('Dataset has been split into train, test, and val sets.')
```

```python
import cv2

from ultralytics import YOLO

# Load the YOLO model
model = YOLO("./models/trained_model_tuned.pt")

# Open the video file
video_path = "reality_test/test2.mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLO inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLO Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
